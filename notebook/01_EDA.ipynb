```python
df = df.dropna()

df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"], format="%Y-%m-%d")

df.head()
```

```python
df.info()
```

```python
df["Dt_Customer"].describe()
```

```python
df["DaysSinceEnrolled"] = df["Dt_Customer"].max() - df["Dt_Customer"]
df["DaysSinceEnrolled"] = df["DaysSinceEnrolled"].dt.days

df.head()
```

```python
df["YearsSinceEnrolled"] = df["DaysSinceEnrolled"] //365

df.head()
```

```python
df["YearsSinceEnrolled"].value_counts()
```

```python
df["Age"] = df["Dt_Customer"].max().year - df["Year_Birth"]

df.head()
```

```python
df.Age.describe()
```

```python
df["AgeGroup"]= pd.cut(
    df["Age"],
    bins = [18,30, 45, 60, df["Age"].max()],
    labels=["18-30","31-45","45-60","61+"],
    include_lowest=True,
)
```

```python
df.head(2)
```

```python
df["AgeGroup"].value_counts()
```

```python
df.columns[df.columns.str.startswith("Mnt")]
```

```python
df.loc[:,df.columns.str.startswith("Mnt")]
```

```python
df["MntTotal"] = (
    df["MntWines"]
    + df["MntFruits"]
    + df["MntMeatProducts"]
    + df["MntFishProducts"]
    + df["MntSweetProducts"]
    + df["MntGoldProds"]
)

df["MntRegularProds"] = df["MntTotal"] - df["MntGoldProds"]

df.head()
```

```python
df["Children"] = df["Kidhome"] + df["Teenhome"]
```

```python
df.loc[df["Children"] == 0,"HasChildren" ] = 0
```

```python
df.loc[df["Children"] != 0,"HasChildren" ] = 1
```

```python
df["HasChildren"].value_counts()
```

```python
df["Marital_Status"] = df["Marital_Status"].replace(
    {
        "Alone":"Single",
        "YOLO":"Single",
        "Absurd":"Single",
        "Together":"Partner",
        "Married":"Partner",
        "Widow":"Single",
        "Single":"Single",
        "Divorced":"Single",
        "Divorced":"Single"
    }
)

df["Marital_Status"].value_counts()
```

```python
df["AcceptedCmpTotal"] = (
    df["AcceptedCmp1"] +
    df["AcceptedCmp2"] +
    df["AcceptedCmp3"] +
    df["AcceptedCmp4"] +
    df["AcceptedCmp5"]
      )

df.loc[df["AcceptedCmpTotal"]==0, "HasAcceptedCmp"] = 0
df.loc[df["AcceptedCmpTotal"]!=0, "HasAcceptedCmp"] = 1
```

```python
df["AcceptedCmpTotal"].value_counts()
```

```python
df["HasAcceptedCmp"].value_counts()
```

```python
df.loc[:,df.columns.str.startswith("Num")]
```

```python
df["NumTotalPurchases"] = (
    df["NumWebPurchases"] +
    df["NumCatalogPurchases"] +
    df['NumStorePurchases']
)

df["NumTotalPurchases"].value_counts()
```
